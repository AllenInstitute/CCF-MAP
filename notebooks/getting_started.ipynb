All atlases have three major components: the **template space**, **annotations**, and **ontology**

1. The **template space** is the image volume that defines the orientation, direction, and voxel size. In the macaque case, this is the Mac25 Rhesus template, which is comprised of an averaged T1 weighted MRI from 25 adult rhesus macaques. 
2. The **annotations** are manually defined segmentations of the template space that delineate anatomical structures. 
3. Annotations are named and labeled using the **unified developing human brain annotation (uDHBA) ontology**, which provides taxonomy of the structures and their relationship, along with metadata (eg. label colors)
