All atlases have three major components: the **anatomical template**, **annotation set**, and **parcellation terminology**

1. The **anatomical template** is a reference image volume that defines an **anatomical space**. The anatomical space sets the orientation and direction of the x,y,z axes and the voxel size. In the macaque case, the anatomical template is the Mac25Rhesus template, which is comprised of an averaged T1 weighted MRI from 25 adult rhesus macaques. 
2. The **annotation set** is a set of manually defined segmentations of the template space that delineate anatomical structures. 
3. Annotations are named and labeled using the **Harmonized Ontology of Mammalian Brain Anatomy (HOMBA)**, which provides taxonomy of the structures and their relationship, along with metadata (eg. label colors)

Together, these components define an anatomical atlas.
